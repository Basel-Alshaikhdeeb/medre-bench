# Medical RE Benchmark Results Analysis

This notebook visualizes and compares results from medre-bench experiments.

```python
import json
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_theme(style="whitegrid")

RESULTS_DIR = Path("../outputs")
```

```python
def collect_results(results_dir: Path) -> pd.DataFrame:
    """Collect all metrics.json files into a DataFrame."""
    records = []
    for metrics_file in results_dir.rglob("metrics.json"):
        with open(metrics_file) as f:
            data = json.load(f)
        config = data.get("config", {})
        eval_m = data.get("eval", {})
        records.append({
            "model": config.get("model", "unknown"),
            "dataset": config.get("dataset", "unknown"),
            "seed": config.get("seed", 0),
            "micro_f1": eval_m.get("eval_micro_f1", eval_m.get("micro_f1", 0)),
            "macro_f1": eval_m.get("eval_macro_f1", eval_m.get("macro_f1", 0)),
            "accuracy": eval_m.get("eval_accuracy", eval_m.get("accuracy", 0)),
        })
    return pd.DataFrame(records)

df = collect_results(RESULTS_DIR)
print(f"Found {len(df)} experiment runs")
df.head()
```

```python
# Pivot table: model x dataset (Micro F1, mean +/- std)
pivot = df.pivot_table(
    index="model", columns="dataset", values="micro_f1",
    aggfunc=["mean", "std"]
)
pivot
```

```python
# Heatmap of Micro F1 scores
mean_pivot = df.pivot_table(
    index="model", columns="dataset", values="micro_f1", aggfunc="mean"
)

fig, ax = plt.subplots(figsize=(10, 6))
sns.heatmap(mean_pivot, annot=True, fmt=".3f", cmap="YlOrRd", ax=ax)
ax.set_title("Micro F1 Scores: Model x Dataset")
plt.tight_layout()
plt.savefig("micro_f1_heatmap.png", dpi=150)
plt.show()
```

```python
# Grouped bar chart per dataset
fig, axes = plt.subplots(1, len(df["dataset"].unique()), figsize=(5 * len(df["dataset"].unique()), 5), sharey=True)
if not hasattr(axes, "__len__"):
    axes = [axes]

for ax, dataset in zip(axes, sorted(df["dataset"].unique())):
    subset = df[df["dataset"] == dataset]
    means = subset.groupby("model")["micro_f1"].mean().sort_values(ascending=False)
    stds = subset.groupby("model")["micro_f1"].std().reindex(means.index).fillna(0)
    
    means.plot(kind="bar", ax=ax, yerr=stds, capsize=3, color=sns.color_palette("husl", len(means)))
    ax.set_title(dataset)
    ax.set_ylabel("Micro F1")
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")

plt.tight_layout()
plt.savefig("per_dataset_comparison.png", dpi=150)
plt.show()
```

```python
# Macro F1 vs Micro F1 scatter
fig, ax = plt.subplots(figsize=(8, 6))
for model in df["model"].unique():
    subset = df[df["model"] == model]
    ax.scatter(subset["micro_f1"], subset["macro_f1"], label=model, s=80, alpha=0.7)

ax.plot([0, 1], [0, 1], "--", color="gray", alpha=0.5)
ax.set_xlabel("Micro F1")
ax.set_ylabel("Macro F1")
ax.set_title("Micro F1 vs Macro F1")
ax.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.savefig("micro_vs_macro_f1.png", dpi=150)
plt.show()
```
